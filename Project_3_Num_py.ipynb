1. **Create an array of numbers from 1 to 20 and reshape it into a 4×5 matrix**

```python

import numpy as np

arr = np.arange(1, 21).reshape(4, 5)
print(arr)
```

**2. Create an array of 10 zeros and replace the 5th element with 100**

```python

arr = np.zeros(10, dtype=int)
arr[4] = 100
print(arr)
```

**4.Generate a 1D array of 15 random integers between 1 and 100 and find its min, max, mean, and standard deviation**

```python

arr = np.random.randint(1, 101, 15)
print("Array:", arr)
print("Min:", arr.min())
print("Max:", arr.max())
print("Mean:", arr.mean())
```

**5. Given two arrays of sales and costs, calculate profit and profit percentage arrays**

```python

sales = np.array([200, 300, 400, 500])
costs = np.array([150, 200, 250, 300])

profit = sales - costs
profit_percentage = (profit / sales) * 100

print("Profit:", profit)
print("Profit %:", profit_percentage)
```

**6. Create a 6×6 random matrix and compute row sums, column sums, and total sum**

```python

matrix = np.random.randint(1, 100, (6, 6))
row_sums = matrix.sum(axis=1)
col_sums = matrix.sum(axis=0)
total_sum = matrix.sum()

print("Matrix:\n", matrix)
print("Row sums:", row_sums)
print("Column sums:", col_sums)
print("Total sum:", total_sum)
```

**7. Given stock prices, compute daily returns using vectorized NumPy operations**

```python

prices = np.array([100, 102, 101, 105, 110])
returns = (prices[1:] - prices[:-1]) / prices[:-1]

print("Daily Returns:", returns)
```

**8. Convert an array of annual interest rates into monthly rates using vectorization**

```python

annual_rates = np.array([0.06, 0.08, 0.10])
monthly_rates = annual_rates / 12

print("Monthly Rates:", monthly_rates)
```

**9. Normalize a given array of portfolio weights so that the sum becomes 1**

```python

weights = np.array([0.2, 0.3, 0.4, 0.5])
normalized_weights = weights / weights.sum()

print("Normalized Weights:", normalized_weights)
```

**10. Create a Pandas Series and find mean, max, and index of max value**

```python

import pandas as pd

data = [1200, 1500, 1800, 2000, 2200]
series = pd.Series(data)

print("Mean:", series.mean())
print("Max:", series.max())
print("Index of Max:", series.idxmax())
```
