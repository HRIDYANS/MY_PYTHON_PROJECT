Create a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
ad_adult['income_binary']=np.where(ad_adult['income']=='>50K',1,0)
```

```python
ad_adult.head()
```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
conditions = [
    ad_adult['age'] < 30,
    (ad_adult['age'] >= 30) & (ad_adult['age'] <= 60),
    ad_adult['age'] > 60
]
choices = ['Youth', 'Adult', 'Senior']
ad_adult['age_group'] = np.select(conditions, choices, default='Unknown')
```

Rename the 'educational-num' column to 'education_num'.

```python
ad_adult.rename(columns={'educational-num': 'education_num'}, inplace=True)
```

Replace the missing values represented by '?' in 'workclass' and 'occupation' columns with 'Unknown'.

```python
ad_adult['workclass'] = ad_adult['workclass'].replace('?', 'Unknown')
ad_adult['occupation'] = ad_adult['occupation'].replace('?', 'Unknown')
```

Calculate the average age for each workclass.

```python
average_age_by_workclass = ad_adult.groupby('workclass')['age'].mean()
display(average_age_by_workclass)
```

Find the total capital-gain for each gender.

```python
total_capital_gain_by_gender = ad_adult.groupby('gender')['capital-gain'].sum()
display(total_capital_gain_by_gender)
```

Get the count of each marital-status category.

```python
marital_status_counts = ad_adult['marital-status'].value_counts()
display(marital_status_counts)
```

Find the mean hours-per-week worked by individuals in different education levels.

```python
mean_hours_per_week_by_education = ad_adult.groupby('education')['hours-per-week'].mean()
display(mean_hours_per_week_by_education)
```

Calculate the mean, median, and standard deviation of 'fnlwgt' for each race.

```python
fnlwgt_stats_by_race = ad_adult.groupby('race')['fnlwgt'].agg(['mean', 'median', 'std'])
display(fnlwgt_stats_by_race)
```

Get the unique values and their counts for the 'native-country' column

```python
native_country_counts = ad_adult['native-country'].value_counts()
display(native_country_counts)
```

Filter the dataset to include only individuals with income >50K.

```python
ad_income_gt_50K = ad_adult[ad_adult['income'] == '>50K']
display(ad_income_gt_50K.head())
```

Sort the dataset by age in descending order.

```python
ad_adult_sorted_by_age = ad_adult.sort_values(by='age', ascending=False)
display(ad_adult_sorted_by_age.head())
```

Filter the dataset to include only individuals who work more than 40 hours per week.



```python
ad_high_hours = ad_adult[ad_adult['hours-per-week'] > 40]
display(ad_high_hours.head())
```

Calculate the percentage of missing values in each column.

```python
missing_percentage = (ad_adult.isnull().sum() / len(ad_adult)) * 100
display(missing_percentage.sort_values(ascending=False))
```

Calculate the median income by education level and gender.

```python
median_income_by_education_gender = ad_adult.groupby(['education', 'gender'])['income_binary'].median()
display(median_income_by_education_gender)
```

Find the top 3 occupations with the highest average capital-gain.

```python
top_3_occupations_capital_gain = ad_adult.groupby('occupation')['capital-gain'].mean().sort_values(ascending=False).head(3)
display(top_3_occupations_capital_gain)
```
